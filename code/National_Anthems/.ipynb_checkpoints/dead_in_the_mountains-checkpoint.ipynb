# National anthems - a comparison

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("/Users/peterjost/Downloads/studio-projects/code/National Anthems/National_Anthems.csv")
df
```

```python
df.dtypes
```

```python
df['topic'].value_counts().plot(kind='bar', color='darkblue', title='The Europeans sing about their countries and battles')
```

```python
df['text_sex'].value_counts().plot(kind='bar', color='darkblue', title='Only one anthem is texted by a female writer')
```

```python
df_new = df.set_index('country')
```

```python
df_new.sort_values(by='words', ascending=False).plot(y='words', kind='bar', color='darkblue', title='The many words of the Marseillaise')  
```

```python
df_new.plot(y='year', style='o', title='Many anthems were introduced around 1990')
```
